## data preparation


### data cleaning

```python
df = df.drop(['Unnamed: 4','Unnamed: 5','Unnamed: 6','Unnamed: 7','Unnamed: 8','Unnamed: 9','Unnamed: 10','Unnamed: 11','Unnamed: 12','Unnamed: 13','Unnamed: 14','Unnamed: 15','Unnamed: 16','Unnamed: 17','Unnamed: 18','Unnamed: 19','Unnamed: 20','Unnamed: 21','Unnamed: 22','Unnamed: 23','Unnamed: 24','Unnamed: 25','Unnamed: 26','Unnamed: 27','Unnamed: 28','Unnamed: 29',], axis=1)
df
```

### removing special characters and emojis

```python
import re
df['COMMENT'] = df['COMMENT'].str.replace(r'@',"")
df['COMMENT'] = df['COMMENT'].str.replace(r'#',"")
df['COMMENT'] = df['COMMENT'].str.replace(r'http\S+',"")
df['COMMENT'] = df['COMMENT'].str.replace(r'[^a-zA-Z0-9\s]',"")
df['COMMENT'] = df['COMMENT'].str.replace(r'[" u"\U0001F600-\U0001F64F"]',"")
df['COMMENT'] = df['COMMENT'].str.replace(u"\U0001F300-\U0001F5FF"
        u"\U0001F680-\U0001F6FF"
        u"\U0001F1E0-\U0001F1FF"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        "]+","")
df['COMMENT'] = df['COMMENT'].str.replace(r'[^\w\s]', '')
df['COMMENT'] = df['COMMENT'].str.replace(r'⭐️',"")
df['COMMENT'] = df['COMMENT'].str.replace(r'💙',"")
df
```

### spliting data into x and y

```python
y = df['SENTIMENT']
y
```

```python
x= df['COMMENT']
x
```

### splitting data for training

```python
from re import X
from sklearn.model_selection import train_test_split

# we create four random variables below
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)
X_train
```

## **model building**

###random forest

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
rf = RandomForestClassifier()
rf.fit(X_train_tfidf, y_train)
```

## Predict on the test data

```python
y_pred = rf.predict(X_test_tfidf)
y_pred
```

## accuracy of the model

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(f"{accuracy * 100}")
```

##checkin for a new comment

```python
new_comment = "I love playing esports with friends! 😍"
# Assign just the new comment string, not the entire column
cleaned_new_comment = new_comment
new_comment_tfidf = vectorizer.transform([cleaned_new_comment])
predicted_sentiment = rf.predict(new_comment_tfidf)
print(f"Predicted sentiment: {predicted_sentiment}")
```
